```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y="EdLevel", order=df["EdLevel"].value_counts().index, hue="Age")
plt.title("Distribution of Education Levels")
plt.xlabel("Count")
plt.ylabel("Education Level")
plt.show()
```

```python
df['LearnCode_Sorted'] = df['LearnCode'].apply(
    lambda x: ';'.join(sorted(str(x).split(';'))) if pd.notna(x) else x
)
```

```python
dummies = df['LearnCode'].str.get_dummies(sep=';')
dummies.head()
df = pd.concat([df, dummies], axis=1)
df.head()
```

```python
plt.figure(figsize=(10, 6))
df.rename(columns={"School (i.e., University, College, etc)": "School"}, inplace=True)
sns.countplot(data=df, y="EdLevel", order=df["EdLevel"].value_counts().index, hue="School")
              der=df["EdLevel"].value_counts().index, hue="School (i.e., University, College, etc)")
```
