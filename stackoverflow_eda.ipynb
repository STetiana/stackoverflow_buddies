```python
def find_outliers(group):
    Q1 = group["WorkExp"].quantile(0.25)
    Q3 = group["WorkExp"].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    return group[(group["WorkExp"] < lower) | (group["WorkExp"] > upper)]
```

```python
outliers_by_age = df_eda.groupby("Age").apply(find_outliers)
```

## Conclusion

The **Stack Overflow Developer Survey 2025** pulled in 49,191 responses across 172 variables.

---

### 1. Who Took the Survey?

Mostly working developers. About **76% are professional developers**, with another ~10% who code as part of their job or studies, and ~9% students. 

---

### 2. Age

**25-34 year-olds at 33.6%**. The 35-44 group (26.9%) is right behind them. Entry-level folks (18-24) make up about 18.7%, and developers 55+ are a small but growing presence (~7%).

Long story short: tech is still a young-ish field.

---

### 3. Education: Degrees Are Common, But Not Required

**Bachelor's (41.2%)** and **Master's (25.6%)** are the most common. Plenty of devs took non-traditional routes — bootcamps, self-teaching, some college but no degree. The field has historically been pretty welcoming to people who didn't follow the "standard" path.


---

### 4. Jobs: Mostly Employed

Around **68% are full-time employees**, and about **13.6% are freelancers or independent contractors**. Students account for ~9%, with some overlap with part-timers.

---

### 5. Side Stuff

About **half of employed developers** have something else going on — caregiving, volunteering, side projects. Students are similar, with ~50% also holding jobs or volunteer roles.

---

### 6. Experience: Mostly Under 15 Years, With Some Wild Outliers

Experience is right-skewed — the median is probably somewhere in the **5-10 year range**, and most devs haven't hit 15 years yet. That said, there are some surprisingly high-experience folks in the 18-24 and 25-34 buckets.

---

### 7. How Developers Learn: Mostly Online

Tech docs, online forums, search engines are the top resources — each used by 18k-22k+ respondents.

- **Newcomers and students** lean on courses and videos
- **Experienced devs** gravitate toward docs and community answers (Stack Overflow, forums)
- Formal classroom learning barely shows up in the responses

---

### 8. Salaries: Big Range, Clear Patterns

The salary data covers about **24,000 respondents** who reported compensation, all converted to USD yearly. Here's what stands out:

**Overall distribution:**
- Median salary: **~$75,000/year**
- Mean is higher at ~$102,000 — pulled up by high earners
- The middle 50% of respondents earn between **$38k and $121k**

**Age matters a lot:**
- 18–24 year olds: median ~$21,000 (entry level, makes sense)
- 25–34: jumps to ~$63,000
- 35–44: ~$93,000
- 45–54: ~$102,000
- 55–64: peaks at ~$107,000
- Salary grows steadily with age — experience pays off, literally.

**Education helps, but not as dramatically as you'd expect:**
- Professional degrees (PhD, JD, MD): ~$87,000
- Master's: ~$80,000
- Bachelor's: ~$76,000
- The gap between a degree and no degree is real but not massive — self-taught devs can absolutely compete.

**Where you work matters more than almost anything:**
- US median: **$150,000** — by far the highest
- Switzerland ($143k) and Israel ($141k) are close
- Western Europe (Denmark $99k, UK $95k, Norway $89k) and Australia ($98k) are solid
- Most of the world is nowhere near these numbers

**Remote work pays more:**
- Fully remote: **$88,500**
- Hybrid (flexibility-leaning): **$81,000**
- Hybrid (office-leaning): **$73,000**
- Fully in-person: **$46,400** — the lowest by a wide margin (partly geography — many in-person roles are in lower-wage countries)

**Top-paying roles:**
- Senior exec / C-suite: **$139,000**
- Engineering manager: **$130,000**
- Product manager: **$104,000**
- Cloud infra / Architect / Cybersecurity: $99k–$103k range
- AI/ML engineers sit at $87k — maybe lower than expected given the hype

**Work experience vs salary — weak correlation:**
The raw correlation between years of experience and salary is surprisingly low (**r ≈ 0.05**). Age is a much better predictor than raw experience — which suggests that career level, role, and location override pure time-in-field.

**Bottom line on salaries:** Geography and role are the biggest salary levers. Experience and education matter, but less than where you live and what you do. Remote work closes some of the gap — but the US/Switzerland/Israel tier is genuinely in a league of its own.
