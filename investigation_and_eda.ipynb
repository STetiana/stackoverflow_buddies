##### Comp Analysis by country and currency 

```python
#creat list of distinct Country and currency pairs
country_currency = df[["Country","Currency"]].drop_duplicates()
print(country_currency)
```

```python
df["Currency"].value_counts()
```

```python
#average converted salary by currency
currency_count = df.groupby("Currency")[["ConvertedCompYearly"]].agg(["mean","max","min","count"]).sort_values(by=("ConvertedCompYearly","count"),ascending=False )
currency_count
```

```python
#get the list of currencies that have more than 200 responses
top_currency_list = []
for i in currency_count.index:
    if currency_count.loc[i,("ConvertedCompYearly","count")] > 200:
        top_currency_list.append(i)
print(top_currency_list)

#create a new dataframe with only the top currencies
df_top_currency = df[df["Currency"].isin(top_currency_list)].copy()
df_top_currency
```

```python
sns.barplot(x="Currency", y="ConvertedCompYearly", data=df_top_currency, hue="RemoteWork")
plt.xticks(rotation=90)
plt.show()
```

```python
#average converted salary by country
country_count = df.groupby("Country")[["ConvertedCompYearly"]].agg(["mean","max","min","count"]).sort_values(by=("ConvertedCompYearly","count"),ascending=False )
country_count
```

```python
#get the list of countries that have more than 200 responses
top_country_list = []
for i in country_count.index:
    if country_count.loc[i,("ConvertedCompYearly","count")] > 200:
        top_country_list.append(i)
print(top_country_list)

#create a new dataframe with only the top countries
df_top_country = df[df["Country"].isin(top_country_list)].copy()
df_top_country

```

```python
sns.barplot(x="Country", y="ConvertedCompYearly", data=df_top_country, hue="RemoteWork")
plt.xticks(rotation=90)
plt.show()
```

##### LearnCode and Tools used

```python
category = "LearnCode"
learn_code_df = df[['DevType','LearnCode','ConvertedCompYearly']].copy()
#Split learning sources
learn_code_df['LearnCode'] = learn_code_df['LearnCode'].str.split(';')

# Explode into multiple rows
lc_expanded = learn_code_df.explode('LearnCode', ignore_index=True)

by_category = lc_expanded.groupby([category])[category].count().reset_index(name='Count')
by_category.sort_values(by='Count', ascending=False, inplace=True)

sns.barplot(data=by_category, y=category, x="Count", color='teal')
by_category
```

```python
lc_expanded.groupby("LearnCode")[["ConvertedCompYearly"]].mean().sort_values(by="ConvertedCompYearly", ascending=False)
```

```python
category = "LanguageHaveWorkedWith"
category_df = df[['DevType',category,'ConvertedCompYearly']].copy()
#Split column value
category_df[category] = category_df[category].str.split(';')

# Explode into multiple rows
c_expanded = category_df.explode(category, ignore_index=True)

by_category = c_expanded.groupby([category])[category].count().reset_index(name='Count')
by_category.sort_values(by='Count', ascending=False, inplace=True)

sns.barplot(data=by_category.head(20), y=category, x="Count", color='teal')
by_category
```
