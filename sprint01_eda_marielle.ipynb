### Cleaning / Removal Anomalies

#### ConvertedCompYearly
* Remove rows with comp >= 500000 and <= 3 years experience

```python
#reset df to original state for testing purposes
#df = pd.read_csv("survey_results_public.csv")
```

```python
#get rows with ConvertedCompYearly >= 500000 and WorkExp <= 3 and remove them from the original dataframe
high_pay_low_exp_df = df[(df["ConvertedCompYearly"] >= 500000) & (df["WorkExp"] <= 3)]

#get rows with ConvertedCompYearly higher than 1000000 and remove them from the original dataframe
pay_over_1m_df = df[(df["ConvertedCompYearly"] >= 1000000)]
anomalies_df = pd.concat([high_pay_low_exp_df, pay_over_1m_df])

```

```python
df = df.drop(anomalies_df.index)

```

#### WorkExp
* Remove rows where declared years of experience exceeds maximum possible age

```python
#create column with maximum possible years of experience based on age provided
df["max_age"] = df["Age"].apply(lambda x: int(x.split("-")[1].split(" ")[0]) if "-" in x else int(x.split(" ")[0]) if x != "Prefer not to say" else -1)
df[["Age", "max_age"]]
```

```python
#Drop rows where WorkExp is higher than max_age
df = df[(df["WorkExp"] <= df["max_age"]) | (df["max_age"] == -1)]
df.reset_index(drop=True, inplace=True)
```

### Comp Analysis by country and currency 
* Check in which currency and country people earn the most



#### All countries and currencies

```python
df.groupby("Country")[["ConvertedCompYearly"]].agg(["median","mean","count"]).sort_values(by=("ConvertedCompYearly","median"), ascending=False).round(2)
```

#### Countries and currencies with at least 200 respondents

```python
#get the list of currencies that have more than 200 responses
top_currency_list = []
currency_count = df["Currency"].value_counts()
for i in currency_count.index:
    if currency_count.loc[i] > 200:
        top_currency_list.append(i)
print(top_currency_list)

#create a new dataframe with only the top currencies
df_top_currency = df[df["Currency"].isin(top_currency_list)].copy()
df_top_currency.groupby("Currency")[["ConvertedCompYearly"]].agg(["median","mean","count"]).sort_values(by=("ConvertedCompYearly","median"), ascending=False).round(2)

```

```python
sns.barplot(x="Currency", y="ConvertedCompYearly", data=df_top_currency, hue="RemoteWork")
plt.xticks(rotation=90)
plt.show()
```

```python
#get the list of countries that have more than 200 responses
top_country_list = []
country_count = df["Country"].value_counts()
for i in country_count.index:
    if country_count.loc[i] > 200:
        top_country_list.append(i)
print(top_country_list)

#create a new dataframe with only the top countries
df_top_country = df[df["Country"].isin(top_country_list)].copy()
df_top_country.groupby("Country")[["ConvertedCompYearly"]].agg(["median","mean","count"]).sort_values(by=("ConvertedCompYearly","median"), ascending=False).round(2)


```

```python
sns.barplot(x="Country", y="ConvertedCompYearly", data=df_top_country, hue="RemoteWork")
plt.xticks(rotation=90)
plt.show()
```

### Source of learning and Tools used

#### LearnCode

```python
category = "LearnCode"
learn_code_df = df[['DevType','LearnCode','ConvertedCompYearly']].copy()
#Split learning sources
learn_code_df['LearnCode'] = learn_code_df['LearnCode'].str.split(';')

# Explode into multiple rows
lc_expanded = learn_code_df.explode('LearnCode', ignore_index=True)

by_category = lc_expanded.groupby([category])[category].count().reset_index(name='Count')
by_category.sort_values(by='Count', ascending=False, inplace=True)

sns.barplot(data=by_category, y=category, x="Count", color='teal')
by_category
```
